```python
import matplotlib.pyplot as plt 
import pandas as pd
from pandas_ods_reader import read_ods
```

```python
#------------------VALDIR'S DATA-------------------------

path = '/home/ratchanah/Desktop/thesis/Valdir_data/Valdir_paleo_data.xls'   # finding the file directory
data = pd.read_excel(path, skiprows = 3)                                    # reading the data and skipping the 4 rows of string before the values

#JAR2
age_JAR2   = data.iloc[:,0]   # x-axis reading the value of row 5 from the beginning till the end of column 0
d13C_JAR2  = data.iloc[:,1]
d18O_JAR2  = data.iloc[:,2]   # y-axis reading the value of row 5 from the beginning till the end of column 2
#JAR7
age_JAR7   = data.iloc[:,4]
d13C_JAR7  = data.iloc[:,5]
d18O_JAR7  = data.iloc[:,6]
#JAR4
age_JAR4   = data.iloc[:,8]
d13C_JAR4  = data.iloc[:,9]
d18O_JAR4  = data.iloc[:,10]
#DBE50
age_DBE50  = data.iloc[:,12]
d13C_DBE50 = data.iloc[:,13]
d18O_DBE50 = data.iloc[:,14]
#CER21
age_CER21  = data.iloc[:,16]
d13C_CER21 = data.iloc[:,17]
d18O_CER21 = data.iloc[:,18]
#CER22
age_CER22  = data.iloc[:,20]
d13C_CER22 = data.iloc[:,21]
d18O_CER22 = data.iloc[:,22]


#----------------RATCHANAH'S DATA---------------------------

dbe1_path   = '/home/ratchanah/Desktop/thesis/Ratchanah_data/DBE1_interpolation.ods'
sheet_index = 1
skiprows    = 1
dbe1_data   = read_ods(dbe1_path, sheet_index, skiprows)

#DBE1
age_DBE1  = dbe1_data.iloc[1:,0]
d13C_DBE1 = dbe1_data.iloc[1:,1]
d18O_DBE1 = dbe1_data.iloc[1:,2]
```

```python
#-------------------------------------SOUTHERN PANTANAL (SEPARATED)----------------------------------------------------

#----------------------------------------13C---------------------------------------------------------------

fig1,ax1 = plt.subplots(figsize=(8, 4))    
fig1.patch.set_visible(False)              # Hide the figure's outer border

# Adjust the spines to keep the axes visible but remove the figure's borders
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
# ax1.spines['bottom'].set_visible(False)

ax1.plot(age_JAR2, d13C_JAR2, color = 'blue', label = 'JAR2', linewidth = 1)
ax1.plot(age_JAR7, d13C_JAR7, color = 'orange', label = 'JAR7', linewidth = 1)
ax1.plot(age_JAR4, d13C_JAR4, color = 'green', label = 'JAR4', linewidth = 1)
ax1.plot(age_DBE50, d13C_DBE50, color = 'red', label = 'DBE50', linewidth = 1)
ax1.set_xlabel('age (years BP)')
ax1.set_ylabel(r'$\delta^{13}C$ (‰)' + '\n' + 'stalagmites')
ax1.legend(loc='upper right')
ax1.set_xlim(0,10000)
plt.savefig('/home/ratchanah/Desktop/thesis/Ratchanah_plots/Proxy/13C_Southern_Pantanal - without DBE1.png')


fig2,ax2 = plt.subplots(figsize=(8, 4))    
fig2.patch.set_visible(False)              # Hide the figure's outer border

# Adjust the spines to keep the axes visible but remove the figure's borders
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)

ax2.plot(age_JAR2, d13C_JAR2, color = 'blue', label = 'JAR2', linewidth = 1)
ax2.plot(age_JAR7, d13C_JAR7, color = 'orange', label = 'JAR7', linewidth = 1)
ax2.plot(age_JAR4, d13C_JAR4, color = 'green', label = 'JAR4', linewidth = 1)
ax2.plot(age_DBE50, d13C_DBE50, color = 'red', label = 'DBE50', linewidth = 1)
ax2.plot(age_DBE1, d13C_DBE1, color = 'magenta', label = 'DBE1', linewidth = 1)
ax2.set_xlabel('age (years BP)')
ax2.set_ylabel(r'$\delta^{13}C$ (‰)' + '\n' + 'stalagmites')
ax2.legend(loc='upper right', fontsize='small')
ax2.set_xlim(0,10000)

#plt.title('Paleoclimate Reconstruction During The Holocene Using $\delta^{18}O$ From Stalagmites')
plt.savefig('/home/ratchanah/Desktop/thesis/Ratchanah_plots/Proxy/13C_Southern_Pantanal - with DBE1.png')

#------------------------------------------------18O------------------------------------------------------------------

fig1,ax1 = plt.subplots(figsize=(8, 4))    
fig1.patch.set_visible(False)              # Hide the figure's outer border

# Adjust the spines to keep the axes visible but remove the figure's borders
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
# ax1.spines['bottom'].set_visible(False)

ax1.plot(age_JAR2, d18O_JAR2, color = 'blue', label = 'JAR2', linewidth = 1)
ax1.plot(age_JAR7, d18O_JAR7, color = 'orange', label = 'JAR7', linewidth = 1)
ax1.plot(age_JAR4, d18O_JAR4, color = 'green', label = 'JAR4', linewidth = 1)
ax1.plot(age_DBE50, d18O_DBE50, color = 'red', label = 'DBE50', linewidth = 1)
ax1.set_xlabel('age (years BP)')
ax1.set_ylabel(r'$\delta^{18}O$ (‰)' + '\n' + 'stalagmites')
ax1.legend(loc='lower left')
ax1.set_xlim(0,10000)
plt.savefig('/home/ratchanah/Desktop/thesis/Ratchanah_plots/Proxy/18O_Southern_Pantanal - without DBE1.png')


fig2,ax2 = plt.subplots(figsize=(8, 4))    
fig2.patch.set_visible(False)              # Hide the figure's outer border

# Adjust the spines to keep the axes visible but remove the figure's borders
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)

ax2.plot(age_JAR2, d18O_JAR2, color = 'blue', label = 'JAR2', linewidth = 1)
ax2.plot(age_JAR7, d18O_JAR7, color = 'orange', label = 'JAR7', linewidth = 1)
ax2.plot(age_JAR4, d18O_JAR4, color = 'green', label = 'JAR4', linewidth = 1)
ax2.plot(age_DBE50, d18O_DBE50, color = 'red', label = 'DBE50', linewidth = 1)
ax2.plot(age_DBE1, d18O_DBE1, color = 'magenta', label = 'DBE1', linewidth = 1)
ax2.set_xlabel('age (years BP)')
ax2.set_ylabel(r'$\delta^{18}O$ (‰)' + '\n' + 'stalagmites')
ax2.legend(loc='lower left', fontsize='small')
ax2.set_xlim(0,10000)


#plt.title('Paleoclimate Reconstruction During The Holocene Using $\delta^{18}O$ From Stalagmites')
plt.savefig('/home/ratchanah/Desktop/thesis/Ratchanah_plots/Proxy/18O_Southern_Pantanal - with DBE1.png')
```
